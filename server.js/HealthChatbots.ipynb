```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("SathwikBalu/Health-model")
model = AutoModelForCausalLM.from_pretrained("SathwikBalu/Health-model")
```

```python
tokenizer = get_chat_template(
    tokenizer,
    chat_template = "phi-3", # Supports zephyr, chatml, mistral, llama, alpaca, vicuna, vicuna_old, unsloth
    mapping = {"role" : "from", "content" : "value", "user" : "human", "assistant" : "gpt"}, # ShareGPT style
)
```

```python
class ModelInput(BaseModel):
    prompt: str
```

```python
def extract_assistant_response(response):
    if response and isinstance(response[0], str):
        full_text = response[0]
        start_marker = "<|assistant|>"
        start_index = full_text.find(start_marker)

        if start_index != -1:
            assistant_text = full_text[start_index + len(start_marker):].strip()
            return assistant_text

    return "No valid assistant response found."
```

```python
@app.post('/chat_d')
def chat_model(input_params: ModelInput):
    prompt = input_params.prompt

    FastLanguageModel.for_inference(model_)  # Enable native 2x faster inference

    messages = [
        {"from": "human", "value": prompt},
    ]
    inputs = tokenizer_.apply_chat_template(
        messages,
        tokenize=True,
        add_generation_prompt=True,
        return_tensors="pt",
    ).to("cuda")

    outputs = model_.generate(input_ids=inputs, max_new_tokens=64, use_cache=True)
    res = tokenizer_.batch_decode(outputs)

    # Extract the assistant's response
    assistant_response = extract_assistant_response(res)

    return {"response": assistant_response}
```

```python
@app.post('/chat_h')
def chat_model(input_params: ModelInput):
    prompt = input_params.prompt

    FastLanguageModel.for_inference(model_)  # Enable native 2x faster inference

    messages = [
        {"from": "human", "value": prompt},
    ]
    inputs = tokenizer.apply_chat_template(
        messages,
        tokenize = True,
        add_generation_prompt = True, # Must add for generation
        return_tensors = "pt",
    ).to("cuda")

    outputs = model.generate(input_ids = inputs, max_new_tokens = 64, use_cache = True)
    res = tokenizer.batch_decode(outputs)

    # Extract the assistant's response
    assistant_response = extract_assistant_response(res)

    return {"response": assistant_response}
```
